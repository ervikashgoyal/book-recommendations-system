```python
x=rb.groupby('user_id').count()['rating'].reset_index()
x=x[x['rating']>200]
good_users= x['user_id']
good_users
```

```python
y=rb.groupby('book_title').count()['rating'].reset_index()
y=y[y['rating']>50]
good_books= y['book_title']
good_books
```

```python
newdata1=rb[rb["book_title"].isin(good_books)]
newdata1
```

```python
final_data=newdata1[newdata1["user_id"].isin(good_users)]
final_data=final_data.reset_index().drop('index',axis=1)
final_data
```

```python
pt= final_data.pivot_table(index='book_title',columns='user_id',values='rating').fillna(0)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similtry= cosine_similarity(pt)
```

```python
similtry
```

```python
similtry.shape
```

```python
def recommend(book):
    index= np.where(pt.index==book)[0][0]
    distances= sorted(list(enumerate(similtry[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in distances:
        print(pt.index[i[0]])
    
```

```python
recommend("1984")
```

```python
recommend('Zoya')
```
